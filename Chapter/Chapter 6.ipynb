## 6.4

```python
m = array([[16, 2, 3, 13], [5, 11, 10, 8], [9, 7, 6, 12], [4, 14, 15, 1]])
r, c = m.shape
r, c
```

```python
m2 = zeros((2*r, 2*c))
m2[::2, ::2] = m
m2
```

```python
a = convolve(m2, array([[0, 0, 0], [0, 1, 1], [0, 1, 1]]), mode="constant")
b = correlate(m2, array([[1, 1, 0], [1, 1, 0], [0, 0, 0]]), mode="constant")
a, b
```

```python
convolve(m2.astype(float), array([[1, 2, 1], [2, 4, 2], [1, 2, 1]]) / 4, mode="constant")
```

```python
ne = array([[1, 1, 0], [1, 1, 0], [0, 0, 0]])
bi = array([[1, 2, 1], [2, 4, 2], [1, 2, 1]]) / 4
bc = array([[1, 4, 6, 4, 1], 
            [4, 16, 24, 16, 4], 
            [6, 24, 35, 24, 6], 
            [4, 16, 24, 16, 4], 
            [1, 4, 6, 4, 1]]) / 64
ne, bi, bc
```

```python
r, c = head.shape
hz = zeros((2*r, 2*c))
hz[::2, ::2] = head
imshow(hz)
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(1, 3, 1)
imshow(correlate(hz, ne))
plt.subplot(1, 3, 2)
imshow(correlate(hz, bi))
plt.subplot(1, 3, 3)
imshow(correlate(hz, bc))
```

## 6.5

meshgrid - https://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html

```python
r = range(-256, 256)
x, y = np.meshgrid(r, r)
z = np.sqrt(x**2 + y**2)
t = 1-((z>254.5) & (z<256))*1
imshow(t, cmap="gray")
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(1, 3, 1)
imshow(rescale(t, 0.25, order=0), cmap="gray")

plt.subplot(1, 3, 2)
imshow(rescale(t, 0.25, order=3), cmap="gray")

#plt.subplot(1, 3, 3)
#t4 = rescale(t, 0.25, order=3) > 0.8
#t5 = img_as_ubyte(t4)
#print(t4.min(), t4.max())
#imshow()
```
