## 7.7 Fourier Transforms of Synthetic Images

### Example 1

```python
a = np.hstack([zeros((256, 128)), ones((256, 128))])
af = fftshift(fft2(a))
afa = fftshow(af, "abs")
afl = fftshow(af, "log")
matplot([a, afa, afl])
```

### Example 2 

```python
a = zeros((256, 256))
a[77:177, 77:177] = 1
af = fftshift(fft2(a))
afl = fftshow(af)
matplot([a, afl])
```

### Example 3

```python
ra = range(256)
x, y = np.meshgrid(ra, ra)
b = (x + y < 329) & (x + y > 182) & (x - y > -67) & (x - y < 73)
bf = fftshift(fft2(b))
bfl = fftshow(bf)
matplot([b, bfl])
```

### Example 4

```python
ar = arange(-128, 128)
x, y = np.meshgrid(ar, ar)
z = np.sqrt(x**2 + y**2)
c = z < 15
cf = fftshift(fft2(c))
cfl = fftshow(cf)
matplot([c, cfl])
```

```python
c2 = 1 / (1 + (z / 15)**2)
c2f = fftshift(fft2(c2))
c2fl = fftshow(c2f)
matplot([c2, c2fl])
```

## 7.8 Filtering in the Frquency Domain

### Low Pass Filtering

```python
cm = imread("cameraman.png")
cf = fftshift(fft2(cm))
cfl = fftshow(cf)
matplot([cm, cfl])
```

```python
cf, c, cfcl, cfcil = fft_filter(cm, "ideal", "low", 15)
matplot([cfcl, cfcil])
```

```python
cf, c, cfcl, cfcil = fft_filter(cm, "ideal", "low", 5)
matplot([cfcl, cfcil])
```

```python
cf, c, cfcl, cfcil = fft_filter(cm, "ideal", "low", 30)
matplot([cfcl, cfcil])
```

### High Pass Filtering

```python
cf, c, cfcl, cfcil = fft_filter(cm, "ideal", "high", 15)
matplot([cfcl, cfcil])
```
